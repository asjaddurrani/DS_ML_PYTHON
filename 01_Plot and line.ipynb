```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
phool=sns.load_dataset("iris")
phool







```

```python
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool Graph")
plt.xlim(1)     # x-axis
plt.ylim(1)     # y-axis
sns.set_style("dark")   #dark or light color
sns.set_style(style=None,rc=None)  # dark or light color
plt.show()
```

### size of figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
phool=sns.load_dataset("iris")

#change figure
plt.figure(figsize=(5,5))

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool Graph")
plt.xlim(1)     # x-axis
plt.ylim(1)     # y-axis
plt.show()
```

### Bar Plot

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
phool=sns.load_dataset("iris")
phool

#draw a line plot
sns.barplot(x="species",y="sepal_width",data=phool)
plt.title("Phool Graph")
plt.show()





```

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
ship=sns.load_dataset("titanic")
ship

```

```python
#draw a line plot
sns.barplot(x="sex",y="age",hue="who",data=ship)
plt.title("Phool Graph")
plt.show()
```

### change order,color and error bars

```python
#draw a line plot
sns.barplot(x="sex",y="age",hue="who",data=ship,order=["female","male"],color="black",ci=None,
           palette='pastel')
plt.title("Phool Graph")
plt.show()
```

```python
### use numpy library for estimator
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load dataset
ship=sns.load_dataset("titanic")
ship

#draw a line plot
sns.barplot(x="sex",y="age",hue="who",data=ship,estimator=np.mean,saturation=0.5)
plt.title("Phool Graph")
plt.show()
```
