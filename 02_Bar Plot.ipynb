###  horizental plot

```python

#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
ship=sns.load_dataset("titanic")
ship

#draw a line plot
sns.barplot(x="age",y="sex",hue="who",data=ship)
plt.title("Phool Graph")
plt.show()
```

###  edeges and borders information

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
ship=sns.load_dataset("titanic")
ship

#draw a line plot
sns.barplot(x="sex",y="age",data=ship,linewidth=3,facecolor=(0.5,0.5,0.5,0.5),errcolor=".5",edgecolor="0")
plt.title("Phool Graph")
plt.show()
```
