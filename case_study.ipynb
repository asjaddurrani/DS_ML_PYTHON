# Pandas Case Study

- We will check the data from titanic data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ship = sns.load_dataset(('titanic'))
ship.head(5)

sns.pointplot(x='sex',y='age',data=ship)
```

- To save a file using line of code

```python
ship.to_csv('ship.csv')
```

```python
# showing basic statistics and summary
ship.describe()
```

```python
ship.head(5)
```

- To remove a single or multiple rows and columns

```python
ship.drop(['deck','sex'],axis=1).head(3)   # we are only fetching the head of data
```

- Total no of value occurence

```python
ship.value_counts(['survived'])
```

- Group on the basis of:

```python
ship.groupby(['sex','class']).mean()
```
